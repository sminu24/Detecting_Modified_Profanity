# 2. Model load

학습을 위해 Pretrained language model (PLM) 사용
> * Hugging face의 ElectraForTokenClassification 사용
> * Data collator로 DataCollatorForTokenClassification 사용

```python
from transformers import TrainingArguments, Trainer, AutoTokenizer, ElectraForTokenClassification, DataCollatorForTokenClassification,pipeline
import torch
import numpy as np
```

```python
model_name = 'beomi/KcELECTRA-base-v2022'
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# 레이블 리스트
fw_labels = ['O', 'B-FW', 'I-FW']

# 레이블을 고유한 숫자 ID로 매핑하기 위한 딕셔너리 생성
label_map = {label: idx for idx, label in enumerate(fw_labels)}
label_map
```

### Model에 사용할 데이터를 전처리
> * text 토크나이징
> * labels 숫자 형태로 변환

```python
def preprocess_function(examples):
    tokenized_examples = tokenizer(str(examples["content"]))

    # str형식의 labels를 숫자형태로 변환
    label_ids = [label_map[label] for label in  eval(examples['labels'])]
    label_ids = [-100]+label_ids  # CLS token
    label_ids.append(-100)  # SEP token
    tokenized_examples['labels'] = torch.tensor(label_ids, dtype=torch.float)


    return tokenized_examples
```

```python
tokenized_dataset = new_dataset.map(preprocess_function)
tokenized_dataset.set_format(type='torch', columns=['input_ids', 'labels', 'attention_mask', 'token_type_ids'])
```

```python
data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer,return_tensors='pt')
num_labels=len(fw_labels) # Label 갯수
```

# 3. Model training

```python
import numpy as np
import evaluate

metric = evaluate.load("seqeval")
def compute_metrics(eval_preds):
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)

    # Remove ignored index (special tokens) and convert to labels
    true_labels = [[fw_labels[l] for l in label if l != -100] for label in labels]
    true_predictions = [
        [fw_labels[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    all_metrics = metric.compute(predictions=true_predictions, references=true_labels)
    return {
        "precision": all_metrics["overall_precision"],
        "recall": all_metrics["overall_recall"],
        "f1": all_metrics["overall_f1"],
        "accuracy": all_metrics["overall_accuracy"],
    }
```

```python
batch_size = 32
```

```python
keys = list(tokenized_dataset.keys())[:-2]
keys
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
for key in keys:
    model = ElectraForTokenClassification.from_pretrained(
    model_name,
    num_labels=num_labels
    )
    model.config.id2label = {i: label for i, label in zip(range(num_labels), fw_labels)}
    model.config.label2id = {label: i for i, label in zip(range(num_labels), fw_labels)}
    print(f"TRAIN {key}!!")
    
    args = TrainingArguments(
        output_dir=f"model_output/{key}",
        evaluation_strategy="epoch",
        learning_rate=1e-4,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        num_train_epochs=10,
        save_strategy='epoch',
        load_best_model_at_end=True,
        greater_is_better=True,
        adam_epsilon=1e-8,
        warmup_ratio=0.1,
        weight_decay = 0.01,
        metric_for_best_model='f1'
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=tokenized_dataset[key],
        eval_dataset=tokenized_dataset["valid"],
        compute_metrics=compute_metrics,
        tokenizer=tokenizer,
        data_collator=data_collator
    )

    trainer.train()
    
    trainer.save_model(f'result/{key}')
```

# 4. Model test

```python
from transformers import TrainingArguments, Trainer, AutoTokenizer, ElectraForTokenClassification, DataCollatorForTokenClassification,pipeline
import torch
import numpy as np
```

```python
# 레이블 리스트
fw_labels = ['O', 'B-FW', 'I-FW']

# 레이블을 고유한 숫자 ID로 매핑하기 위한 딕셔너리 생성
label_map = {label: idx for idx, label in enumerate(fw_labels)}
label_map
```

```python
from sklearn.metrics import f1_score
from tqdm.notebook import tqdm

result = {}
for key in keys:
    print(f'----------{key}----------')
    saved_model_path = f'result/{key}'
    tokenizer = AutoTokenizer.from_pretrained(saved_model_path)

    num_labels=len(fw_labels) # Label 갯수

    model = ElectraForTokenClassification.from_pretrained(
        saved_model_path,
        num_labels=num_labels
    )

    model.config.id2label = {i: label for i, label in zip(range(num_labels), fw_labels)}
    model.config.label2id = {label: i for i, label in zip(range(num_labels), fw_labels)}

    scores = []
    all_predictions,all_true_labels = [],[]
    for example in tqdm(valid_dataset):
        true_label = [['O']+list(eval(example['labels']))+['O']]
        text = example['content']
        inputs = tokenizer(text,return_tensors="pt")
        with torch.no_grad():
            logits = model(**inputs).logits
        predictions = torch.argmax(logits, dim=2)
        predicted_token_class = [[model.config.id2label[t.item()] for t in predictions[0]]]
        all_predictions.extend(predicted_token_class)
        all_true_labels.extend(true_label)
        
    all_metrics = metric.compute(predictions=all_predictions, references=all_true_labels)
    f1_score = all_metrics["overall_f1"]
    result[key] = f1_score
    print(f'f1-score: {f1_score}')
    print()
```
